```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Step 1: Generate Sample Data
np.random.seed(50)
citric_acid = np.random.uniform(1, 5, 200)  # Citric acid content between 1 and 5
density = np.random.uniform(9930, 9970, 200)  # Density between 9930 and 9970

# Step 2: Calculate Quality
quality = 0.0015 * (citric_acid / density)

# Step 3: Categorize Wine Quality
def categorize_quality(q):
    if 5 <= q < 6.5:
        return "normal"
    elif 6.5 <= q < 8:
        return "good"
    else:
        return "best"

quality_labels = np.array([categorize_quality(q) for q in quality])

# Encode the quality labels
label_encoder = LabelEncoder()
quality_encoded = label_encoder.fit_transform(quality_labels)

# Create a DataFrame
wine_df = pd.DataFrame({
    "Citric_Acid": citric_acid,
    "Density": density,
    "Quality": quality,
    "Quality_Label": quality_labels,
    "Quality_Encoded": quality_encoded
})

# Step 4: Plot the Data
plt.figure(figsize=(10, 6))
plt.scatter(wine_df["Citric_Acid"], wine_df["Density"], c=wine_df["Quality_Encoded"], cmap='viridis', s=50)
plt.xlabel("Citric Acid Content")
plt.ylabel("Density")
plt.title("Wine Quality Visualization")
plt.colorbar(ticks=[0, 1, 2], label="Wine Status (0=normal, 1=good, 2=best)")
plt.show()

# Step 5: Train a KNN Model
X = wine_df[["Citric_Acid", "Density"]]
y = wine_df["Quality_Encoded"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)

model = KNeighborsClassifier(n_neighbors=4)
model.fit(X_train, y_train)

# Step 6: Make Predictions and Visualize
y_pred = model.predict(X_test)

plt.figure(figsize=(10, 6))
plt.scatter(X_test["Citric_Acid"], X_test["Density"], c=y_pred, cmap='viridis', s=50)
plt.xlabel("Citric Acid Content")
plt.ylabel("Density")
plt.title("Wine Quality Prediction using KNN")
plt.colorbar(ticks=[0, 1, 2], label="Predicted Wine Status (0=normal, 1=good, 2=best)")
plt.show()

```
