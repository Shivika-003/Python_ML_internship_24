```python
"""Q2. create a program that promotes the user to input two numbers and perform division.
Handle the Arithematics exception that might occur if the user attempts to divide by zero and potential 
NumberFormatExceptin that might occur during the conversion.
IN SUCH CASE ,INFORM THE USER ABOUT THE ERROR AND ASK FOR INPUT AGAIN"""
def get_number(prompt):
    while True:
        try:
            return int (input(prompt))
        except ValueError:
            print("Invalid input.Enter a valid number.")

def divide_numbers():
    while True:
        num1 = get_number("Enter the first num: ")
        num2 = get_number("Enter the second num: ")
        
        try:
            result = num1 / num2
            print(f"The result of {num1} divided by {num2} is {result}")
            break
        except ZeroDivisionError:
            print("Division by zero is not allowed.Enter the numbers again.")

divide_numbers()

```

```python
"""Q3. SPORTS  EQUIPMENT RENTAL SYSTEM POC:
Create a class equipment that stores 
EID, ENAME, BRAND TOTAL, AVAL QTY
Add Equipment to stock 
Delete Equipment from stock 
Update the quantity
Print the list of available equipment
Create a new dictionary that stores the record of students who take the sports equipment and name is a
rental.Dictionary contains:Name , rollno,Eid, Ename,status, return condition
create two functions as rent item and return item
Rent item = this function will generate the new record to the dictionary rental
and fill all the details and mark the condition as good and status as Rented
Return item = this function will mark the status as returned if student if student return the item and also mention the condition as (good or damaged).

explanation of question flow chart type

1. stock managment (is secondary task)
2. issue register (retal register is main task ) 
=>register ->name,id,equipment id ,equipment name[,status{rented, returned},condition{good, bad }]-> situation based

=>equipment { E id,E name,Ebrand,EI qty, EA qty}\function ->add item, rename item, update item (qty), read stock
=>return
=> return"""
class Equipment:
    def __init__(self, eid, ename, brand, total):
        self.eid = eid
        self.ename = ename
        self.brand = brand
        self.total = total
        self.aval_qty = total

    def add_item(self, qty):
        self.total += qty
        self.aval_qty += qty

    def delete_item(self, qty):
        if qty <= self.aval_qty:
            self.total -= qty
            self.aval_qty -= qty
        else:
            print("Not enough available quantity to delete.")

    def update_quantity(self, new_qty):
        self.aval_qty = new_qty

    def print_available(self):
        print(f"EID: {self.eid}, ENAME: {self.ename}, BRAND: {self.brand}, TOTAL: {self.total}, AVAILABLE: {self.aval_qty}")

class RentalSystem:
    def __init__(self):
        self.rental_records = {}

    def rent_item(self, name, rollno, eid, ename):
        if eid in self.rental_records:
            print(f"Equipment with EID {eid} is already rented.")
            return
        record = {
            "Name": name,
            "RollNo": rollno,
            "EID": eid,
            "EName": ename,
            "Status": "Rented",
            "Condition": "Good"
        }
        self.rental_records[eid] = record

    def return_item(self, rollno, eid, condition):
        if eid not in self.rental_records:
            print(f"No rental record found for EID {eid}.")
            return
        record = self.rental_records[eid]
        if record["RollNo"] == rollno:
            record["Status"] = "Returned"
            record["Condition"] = condition
        else:
            print(f"No rental record found for RollNo {rollno} and EID {eid}.")

    def print_rentals(self):
        for record in self.rental_records.values():
            print(record)

# Example Usage
# Create some equipment
bat = Equipment(1, "Cricket Bat", "SS", 10)
ball = Equipment(2, "Cricket Ball", "Kookaburra", 20)

# Print initial stock
bat.print_available()
ball.print_available()

# Update stock
bat.add_item(5)
ball.delete_item(5)

# Print updated stock
bat.print_available()
ball.print_available()

# Create a rental management system
rental_system = RentalSystem()

# Rent items
rental_system.rent_item("John Doe", 101, 1, "Cricket Bat")
rental_system.rent_item("Jane Smith", 102, 2, "Cricket Ball")

# Print rental records
rental_system.print_rentals()

# Return items
rental_system.return_item(101, 1, "Good")
rental_system.return_item(102, 2, "Damaged")

# Print updated rental records
rental_system.print_rentals()

       
```
