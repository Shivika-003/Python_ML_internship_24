```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame for demonstration purposes
data = {
    'survived': [0, 1, 1, 1, 0],
    'pclass': [3, 1, 3, 1, 3],
    'sex': ['male', 'female', 'female', 'female', 'male'],
    'age': [22.0, 38.0, 26.0, 35.0, 35.0],
    'sibsp': [1, 1, 0, 1, 0],
    'parch': [0, 0, 0, 0, 0],
    'fare': [7.25, 71.2833, 7.925, 53.1, 8.05],
    'embarked': ['S', 'C', 'S', 'S', 'S'],
    'class': ['Third', 'First', 'Third', 'First', 'Third'],
    'who': ['man', 'woman', 'woman', 'woman', 'man'],
    'adult_male': [True, False, False, False, True],
    'deck': [None, 'C', None, None, None],
    'embark_town': ['Southampton', 'Cherbourg', 'Southampton', 'Southampton', 'Southampton'],
    'alive': ['no', 'yes', 'yes', 'yes', 'no'],
    'alone': [False, False, True, False, True]
}

df = pd.DataFrame(data)

# Group by 'class' and 'sex', then unstack to create a DataFrame suitable for plotting
class_vs_sex = df.groupby(["class", "sex"]).size().unstack(fill_value=0)

# Plot stacked bar chart
plt.figure(figsize=(12, 8))
class_vs_sex.plot(kind='bar', stacked=True)
plt.title('Class vs Sex (Stacked)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Plot non-stacked bar chart
plt.figure(figsize=(12, 8))
class_vs_sex.plot(kind='bar', stacked=False)
plt.title('Class vs Sex (Non-stacked)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
